```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob

%load_ext autoreload
%autoreload 2
```

```python
py_files = glob.glob("/Volumes/Transcend/programs/prog1/python/*.py")
cpp_files = glob.glob("/Volumes/Transcend/programs/prog1/cpp/*.cpp")

max_samples = 2000
sample_cnt = 0
py_text_list = []
cpp_text_list = []

for f_py, f_cpp in zip(py_files,cpp_files):
    with open(f_py, "rb") as infile:
        py_text_list.append(infile.read().decode("utf-8") )
    with open(f_cpp, "rb") as infile:
        cpp_text_list.append(infile.read().decode("utf-8"))
    sample_cnt += 1
    if sample_cnt == max_samples:
        break
```

```python
df_dict = {'python':py_text_list ,
           'cpp': cpp_text_list}

df = pd.DataFrame.from_dict(df_dict)
df.to_pickle('data.pkl')
```
